# **Plot Notebook**

This notebook contains all the code for the plots in the paper.

```python
import matplotlib.pyplot as plt
```

The first plot is for the sample concept weights in Figure 1.

```python
# Size adjustment factor
factor = 1.1

# Data
categories = ['Irregular Streaks', 'Blue Whitish Veils', 'Canine', 'Ruff', 'Fur', 'Flower', 'Thorns', 'Pink']
values = {
    'HAM10000': [0.333, 0.294],
    'CIFAR100': [13.27, 9.07, 7.35, 24.7, 14.9, 9.84]
}

# Font
font = "Californian FB"

# Colors
ham_colors = ['chartreuse', 'chartreuse']
cifar_colors = ['orange', 'orange', 'orange']

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 4))

# Plot HAM10000 with values on top
bars = axes[0, 0].barh(categories[:2], values['HAM10000'][:2], color=ham_colors)
for bar, value in zip(bars, values['HAM10000'][:2]):
    axes[0, 0].text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{value:.3f}', ha='left', va='center', fontdict={'fontname': font})

# Plot HAM10000 (second subplot) with values on top
bars = axes[1, 0].barh(categories[:2], values['HAM10000'][:2], color=ham_colors)
for bar, value in zip(bars, values['HAM10000'][:2]):
    axes[1, 0].text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{value:.3f}', ha='left', va='center', fontdict={'fontname': font})

# Plot CIFAR100 with values on top
bars = axes[0, 1].barh(categories[2:5], values['CIFAR100'][:3], color=cifar_colors)
for bar, value in zip(bars, values['CIFAR100'][:3]):
    axes[0, 1].text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{value:.3f}', ha='left', va='center', fontdict={'fontname': font})

# Plot CIFAR100 (second subplot) with values on top
bars = axes[1, 1].barh(categories[5:], values['CIFAR100'][3:], color=cifar_colors)
for bar, value in zip(bars, values['CIFAR100'][3:]):
    axes[1, 1].text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{value:.3f}', ha='left', va='center', fontdict={'fontname': font})

# Set titles with custom font style
title_font = {'fontname': font, 'weight': 'bold', 'fontsize':14}
axes[0, 0].set_title('HAM10000', **title_font)
axes[0, 1].set_title('CIFAR100', **title_font)

for ax in axes.flatten():
    ax.set_axisbelow(True)
    ax.grid(axis='x', alpha=0.6)
    ax.tick_params(axis='both', labelsize=10, labelfontfamily=font)

    # Adjust x-axis limits for more space on the sides
    ax.set_xlim(0, max(ax.get_xlim()) * factor)

axes[0, 0].text(1.03, 0.5, 'Benign', ha='center', va='center', rotation=270, transform=axes[0, 0].transAxes, fontdict={'fontname': font, 'fontsize': 12})
axes[1, 0].text(1.03, 0.5, 'Malignant', ha='center', va='center', rotation=270, transform=axes[1, 0].transAxes, fontdict={'fontname': font, 'fontsize': 12})
axes[0, 1].text(1.03, 0.5, 'Dog', ha='center', va='center', rotation=270, transform=axes[0, 1].transAxes, fontdict={'fontname': font, 'fontsize': 12})
axes[1, 1].text(1.03, 0.5, 'Rose', ha='center', va='center', rotation=270, transform=axes[1, 1].transAxes, fontdict={'fontname': font, 'fontsize': 12})

# Invert the order
axes[0, 0].invert_yaxis()
axes[1, 0].invert_yaxis()
axes[0, 1].invert_yaxis()
axes[1, 1].invert_yaxis()

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

The second is for the audio concept weight figure in Figure _ .

```python
# Size adjustment factor
factor = 1.1

# Data
categories = [['Babbling', ' Hubbub,\n speech noise,\n speech babble', 'Female speech,\n woman speaking'],
              ['Growling', 'Yip', 'Bow-wow']]
values = {
    'children': [20.982, 18.338, 15.307],
    'dog': [29.478, 27.129, 21.087]
}

# Font
font = "Californian FB"

# Colors
dog_colors = ['navajowhite', 'navajowhite', 'navajowhite']
child_colors = ['aquamarine', 'aquamarine', 'aquamarine']

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(6, 4))

# Plot Children Playing Class concepts
bars = axes[0].barh(categories[0], values['children'], color=child_colors)
for bar, value in zip(bars, values['children']):
    axes[0].text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{value:.3f}', ha='left', va='center', fontdict={'fontname': font})

# Plot Dog Class concepts
bars = axes[1].barh(categories[1], values['dog'], color=dog_colors)
for bar, value in zip(bars, values['dog']):
    axes[1].text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{value:.3f}', ha='left', va='center', fontdict={'fontname': font})

# Set titles with custom font style
title_font = {'fontname': font, 'weight': 'bold', 'fontsize':14}
axes[0].set_title('UrbanSound8K', **title_font)

for ax in axes.flatten():
    ax.set_axisbelow(True)
    ax.grid(axis='x', alpha=0.6)
    ax.tick_params(axis='both', labelsize=10, labelfontfamily=font)

    # Adjust x-axis limits for more space on the sides
    ax.set_xlim(0, max(ax.get_xlim()) * factor)

axes[0].text(1.03, 0.5, 'Children Playing', ha='center', va='center', rotation=270, transform=axes[0].transAxes, fontdict={'fontname': font, 'fontsize': 12})
axes[1].text(1.03, 0.5, 'Dog Barking', ha='center', va='center', rotation=270, transform=axes[1].transAxes, fontdict={'fontname': font, 'fontsize': 12})

# Invert the order
axes[0].invert_yaxis()
axes[1].invert_yaxis()

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
